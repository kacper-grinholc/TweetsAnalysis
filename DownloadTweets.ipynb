```python
from datetime import date, timedelta
import snscrape.modules.twitter as sntwitter
import re
import datetime
import calendar
```

```python
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
hashtags = ['Ukraine', 'Russia', 'Poland', 'Israel', 'Turkey']
maxTweets = 200
        
start_date = date(2021, 6, 1)
end_date = date(2022, 6, 1)
```

```python
for hashtag in hashtags:
    for single_date in daterange(start_date, end_date):
        day_start = single_date.strftime("%Y-%m-%d")
        day_end = str(datetime.datetime.strptime(day_start, "%Y-%m-%d") + timedelta(days=1))[0:10]
        month = datetime. datetime.strptime(day_start, "%Y-%m-%d").month
        day = datetime. datetime.strptime(day_start, "%Y-%m-%d").day
        month = calendar.month_name[month]
        print(str(day), month, hashtag)
        tweets = []
        for i,tweet in enumerate(sntwitter.TwitterHashtagScraper(hashtag + ' since:' + day_start + ' until:' + day_end + ' lang:en').get_items()) :
                if i >= maxTweets :
                    break
                result = re.sub(r"http\S+", "", tweet.content)
                result = str.join(" ", result.splitlines())
                tweets.append(result)
        with open("Tweets/" + month + "/" + hashtag + "/" + str(day) + ".txt", "x") as txt_file:
            for line in tweets:
                txt_file.write(str(line) + "\n")
```
